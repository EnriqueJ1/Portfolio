# PERFORMING RANDOM SIMULATIONS

```python
#GENERAL POP
#HEALTHY AND UNHEALTHY
#TOTAL POP: 10,000
#"L": Leukemia - 4
#"H": Healthy - 9996
#MAKE UP DATA OF STRINGS
unhealthy = 4*["L"] #leukemia data
healthy = 9996*["H"]#healthy data
pop = unhealthy + healthy #general pop

#ARRAYS
#STORAGE PLACE
import numpy as np

pop_arr = np.array( pop ) #converting a list to an array
```

```python
#TAKE A RANDOM SAMPLE FROM GENERAL POPULATION
#SPECIFICALLY LET'S TAKE 11,748 - TOTAL POP IN WOBURN
random_sample = np.random.choice( pop_arr , 11748 ) #take a random sample of 11,748 from the general population array

#DETERMINE HOW MANY PEOPLE HAVE LEUKEMIA IN THIS RANDOM SAMPLE

leu = random_sample == "L" #determining which people have leukemia

total_leu = np.sum( leu ) #determining the total number of people who have leukemia

print(total_leu) #this result is for only one simulation
```

```python
#PERFORM MULTIPLE SIMULATIONS
#USING A FOR LOOP
#PERFORM 10 SIMULATIONS

simulations = 10 #number of simulations

sample_size = 11748 #size of random sample size

for i in range(simulations): #for each simulation
    random_sample = np.random.choice( pop_arr , sample_size ) #take a random sample from general population
    leu = random_sample == "L" #determining which people have leukemia
    total_leu = np.sum( leu ) #sum up total number of leukemia people for current simulation
    print("Number of leukemia cases for current simulation:",total_leu )
    print("\n") #space line
```

```python
#PERFORM MULTIPLE SIMULATIONS
#USING A FOR LOOP
#WILL STORE LEUKEMIA CASES TO AN EMPTY LIST
#PERFORM 10,000 SIMULATIONS

simulations = 10000 #number of simulations

leu_list = [] #empty list used to store number of people who have leukemia for each simualtion

for i in range(simulations): #for each simulation
    random_sample = np.random.choice( pop_arr , 11748 ) #take a random sample from general population
    leu = random_sample == "L" #determining which people have leukemia
    total_leu = np.sum( leu ) #sum up total number of leukemia people for current simulation
    leu_list.append( total_leu )

print(leu_list) #leu_list should not longer be empty and should contain all the lekemia cases for each simulation
#IT IS NOT NECCESARY TO PRINT THE LIST BUT IT IS IMPORTANT TO UNDERSTAND WHAT THE VALUES IN YOUR LIST MEAN
```

```python
#PLOT HISTOGRAM PLOT
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

p = sns.displot( leu_list )
p.set(xlabel = "Number of people with Leukemia out of 11,748") #labeling x-axis
plt.axvline(12, color="red") #drawing a vertical line at our observed case - 12 cases of leukemia in Woburn
```

```python
#RIGHT AWAY WE CAN SEE THAT 12 CASES OF LEUKEMIA DOES NOT FALL IN THE BULK OF OUR DISTRIBUTION
#LET'S FIND THE p-value ASSOCIATED WHICH SUCH EVENT OCCURING OR A MORE EXTREME CASES OCCURRING
#IN OTHER WORDS LET'S FIND THE PROPABILITY OF 12 OR MORE LEUKEMIA CASES OCCURRING

leu_arr = np.array( leu_list ) #convert list to array
extreme = leu_arr >= 12 #determing which values contains 12 cases or more of leukemia
total_extreme = np.sum(extreme) #adding up all the extreme cases
pvalue = total_extreme/simulations #total extreme cases divided by the total number of cases
print(pvalue)
```

The types of simulations weve been using so far are called Monte Carlo simulations.

Simulate throwing a fair coin 100 and 10,000 times and graph the results.


```python
#LET 1 REPRESENT THE VALUE HEAD
#LET 2 REPRESENT THE VALUE TAIL
#THE BOX MODEL REPRESENT OUR POPULATION/OR POSSIBLE VAALUES
#METHOD 1

import numpy as np
import seaborn as sns
%matplotlib inline

box_model = [1,2] #made a list with equal chance of getting a head or tail

tosses = 100 #number of tosses

random_toss = np.random.choice( box_model , tosses ) #randomly get 100 values from the box_model, which we will interpret each value as a different toss

#PLOT TOSS DISTRIBUTION AS HISTOGRAM PLOT
p = sns.displot( data = random_toss )
p.set( xlabel = 'Number of head or tail outcomes' )

#or

p2 = sns.displot( data = random_toss , bins = 2 ) #there is only two possible cases so the plot will look much better with only two bins
p2.set( xlabel = 'Number of head or tail outcomes' )

#EACH TIME YOU RE-RUN THE CODE, YOU WILL GET SLIGHTLY DIFFERENT DISTRIBUTIONS BECAUSE OF THE RANDOM CHOOSING - np.random.choice
#BUT IF WE PERFORMED MORE TOSSES WE WILL SEE THAT THE NUMBER OF HEAD AND TAIL OUTCOMES GET CLOSER
```

```python
#LET 1 REPRESENT THE VALUE HEAD
#LET 2 REPRESENT THE VALUE TAIL
#THE BOX MODEL REPRESENT OUR POPULATION/OR POSSIBLE VAALUES
#METHOD 2

#THIS METHOD GENERALIZES BETTER TO MULTPLE COINS

import numpy as np
import seaborn as sns
%matplotlib inline

box_model = [1,2] #made a list with equal chance of getting a head or tail

simulations = 100 #in each simulation toss the coin once

toss_list = [ ] #empty list to store all toss values
for i in range( simulations ): #for each coin toss

    random_toss = np.random.choice( box_model , 1 ) #only toss one random coin

    value_of_toss = random_toss[0] #random_toss contains a single value which needs to be extracted

    toss_list.append( value_of_toss ) #store toss value into empty list

#PLOT TOSS DISTRIBUTION AS HISTOGRAM PLOT
p = sns.displot( data = toss_list )
p.set( xlabel = 'Number of head or tail outcomes' )

#or

p2 = sns.displot( data = toss_list , bins = 2 ) #there is only two possible cases so the plot will look much better with only two bins
p2.set( xlabel = 'Number of head or tail outcomes' )

#EACH TIME YOU RE-RUN THE CODE, YOU WILL GET SLIGHTLY DIFFERENT DISTRIBUTIONS BECAUSE OF THE RANDOM CHOOSING - np.random.choice
#BUT IF WE PERFORMED MORE TOSSES WE WILL SEE THAT THE NUMBER OF HEAD AND TAIL OUTCOMES GET CLOSER
```
