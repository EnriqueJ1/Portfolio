```python
%matplotlib inline
import seaborn as sns #importing Seaborn (plotting) library
import numpy as np #importing Numpy library

cars = sns.load_dataset( "mpg" ).dropna() #loading car data from Seaborn library with non-zero entries
cars_data = cars[cars.cylinders.isin( [4,6,8] ) ] #keeping cars with only 4, 6 and 8 cylinders
cars_data
```

```python
#STRIPPLOT OF CAR HORSEPOWER
cars_horsepower = cars_data.horsepower #extracting horsepower data

p = sns.stripplot( data = cars_horsepower ) #plotting dotplot
p.set( ylabel = "Horsepower" ) #labeling y-axis. x-axis has not meaning with this orientation

#MAKE PYTHON FUNCTION TO CALCULATE THE MEAN OF HORSEPOWER DATA
#USE THE MEAN FUNCTION TO CONFIRM
def my_mean( data_list ): #making my own mean function
    return sum( data_list )/len( data_list )

print( my_mean( cars_horsepower ) ) #value of mean with my_mean function
print( np.mean( cars_horsepower ) ) #value of mean with mean function from Numpy
```

```python
#BEESWARM PLOT OF CAR WEIGHT
cars_weight = cars_data.weight #extracting weight data

p2 = sns.swarmplot( data = cars_weight , color = "red", orient = "h" ) #plotting beeswarm dotplot
p2.set( xlabel = "Horsepower" ) #labeling x-axis. y-axis has not meaning with this orientation

#MAKE PYTHON FUNCTION TO CALCULATE THE MEDIAN OF WEIGHT DATA
#USE THE MEDIAN FUNCTION TO CONFIRM
def my_median( data_list ):
    sort_data = sorted( data_list ) #sort data

    #CHECK IF LIST HAS EVEN OR ODD NUMBER OF ELEMENTS
    if len( sort_data )%2 == 0: #checking if the list is divisible by 2
        return (sort_data[ int( len(sort_data)/2 - 1 ) ] + sort_data[ int( len(sort_data)/2 + 1 ) ])/2
    else:
        return sort_data[ int( len(sort_data)/2 ) ]

print( my_median( cars_weight ) ) #value of median with my_median function
print( np.median( cars_weight ) ) #value of median with median function from Numpy
```

```python
#HISTOGRAM PLOT OF CAR MPG
cars_mpg = cars_data.mpg #extracting mpg data

p3 = sns.displot( data = cars_mpg , color = "green" ) #histogram

#MAKE PYTHON FUNCTION TO CALCULATE THE VARIANCE OF MPG DATA
#USE THE VARIANCE FUNCTION TO CONFIRM
def my_var( data_list ):
    x = [] #empty list
    for i in data_list:
        x.append( (i - my_mean(data_list) )**2 )

    return sum(x)/len( data_list )

print("Value of the variance with my_var function")
print( my_var( cars_mpg ) )
print( "\n" )  #spaceline

print( "Value of the variance with variance function from Numpy" )
print( np.var( cars_mpg ) )
```

```python
#BOXPLOT PLOT OF CAR ACCELERATION
cars_acceleration = cars_data.acceleration #extracting acceleration data

p4 = sns.boxplot( data = cars_acceleration , color = "purple", orient = "h" ) #boxplot
p4.set( xlabel = "Acceleration" )

#MAKE PYTHON FUNCTION TO CALCULATE THE STANDARD DEVIATION OF ACCELERATION DATA
#USE THE STD FUNCTION TO CONFIRM
def my_std( data_list ):
    return my_var(data_list)**(1/2) #to raise a number to a power use double asterisk (4**2, 4 raised to the power of 2)
                                    #could also use the np.sqrt() function

print( "Value of the standard deviation with my_std function" )
print( my_std( cars_acceleration ) )
print( "\n" )  #spaceline

print( "Value of the standard deviation with std function from Numpy" )
print( np.std( cars_acceleration ) )
```

```python
#VIOLING PLOT OF CAR DISPLACEMENT YEAR
cars_displacement = cars_data.displacement #extracting displacement data

p5 = sns.violinplot( data = cars_displacement , color = "orange" ) #violin plot
p5.set( ylabel = "Displacement" )

#CALCULATE MEDIAN ABSOLUTE DEVIATION (MAD)
#USING my_median FUNCTION
def MAD( datalist ):
    y = []#empty list
    for i in datalist:
        y.append( abs( i - my_median(datalist) ) )
    return my_median(y)

#CALCULATE MEDIAN ABSOLUTE DEVIATION
#USING MEDIAN FUNCTION IN NUMPY
def MAD2( datalist ):
    z = []#empty list
    for i in datalist:
        z.append( abs( i - np.median(datalist) ) )
    return np.median(z)

print("Value of the MAD with MAD function")
print( MAD(cars_displacement) )
print( "\n" )  #spaceline

print("Value of the MAD with MAD2 function")
print( MAD2(cars_displacement) )
```
